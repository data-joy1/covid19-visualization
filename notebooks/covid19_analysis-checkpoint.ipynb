```python
import pandas as pd

#load dataset
df = pd.read_csv(r"D:\projects for cv\covid19-visualization\data\all_weekly_excess_deaths.csv")

# Preview data
df.head()
```

```python
# Check data types & missing values
df.info()
df.isnull().sum()
```

```python
# Fill missing values with 0
df.fillna(0, inplace=True)
```

```python
# Convert 'date' column to datetime format 
df["start_date"] = pd.to_datetime(df["start_date"])
df["end_date"] = pd.to_datetime(df["end_date"])
df["week"] = pd.to_datetime(df["week"])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_style("darkgrid")
```

```python
plt.figure(figsize=(14, 7))

# Convert 'start_date' column to datetime format if not already done
df["start_date"] = pd.to_datetime(df["start_date"])

# Select only the countries of interest
selected_countries = ["United States", "Australia", "Britain", "Canada"]
df_selected = df[df["country"].isin(selected_countries)]

# Plot each country separately
for country in selected_countries:
    country_data = df_selected[df_selected["country"] == country]
    sns.lineplot(x=country_data["start_date"], y=country_data["covid_deaths"], label=country)

# Formatting
plt.xlabel("Date")
plt.ylabel("Weekly Excess Deaths")
plt.title("Comparison of Weekly Excess Deaths (USA, Australia, UK, Canada)")
plt.xticks(rotation=45)
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc="upper left")

# Save plot
plt.savefig("D:/projects for cv/covid19-visualization/output/covid_deaths_comparison.png", bbox_inches="tight")

# Show plot
plt.show()

```

```python
# Aggregate total deaths by country
top_countries = df.groupby("country")["covid_deaths"].max().sort_values(ascending=False).head(10)

# Bar Plot
plt.figure(figsize=(10, 5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette="Reds_r")

# Formatting
plt.xlabel("Total Deaths")
plt.ylabel("Country")
plt.title("Top 10 Countries with Highest COVID-19 Deaths")

# Save plot
plt.savefig("D:/projects for cv/covid19-visualization/screenshots/top10_deaths.png")

# Show plot
plt.show()
```

```python
# Aggregate total COVID and non-COVID deaths by country
top_countries = df.groupby("country")[["covid_deaths", "non_covid_deaths"]].sum()
top_countries = top_countries.sort_values("covid_deaths", ascending=False).head(10)

# Plot
top_countries.plot(kind="bar", figsize=(12, 6), color=["red", "gray"])

# Formatting
plt.xlabel("Country")
plt.ylabel("Total Deaths")
plt.title("Total COVID Deaths vs. Non-COVID Deaths (Top 10 Countries)")
plt.xticks(rotation=45)
plt.legend(["COVID Deaths", "Non-COVID Deaths"])

# Save plot
plt.savefig("D:/projects for cv/covid19-visualization/screenshots/covid_vs_non_covid_deaths.png")

# Show plot
plt.show()
```

```python
# Select data for USA
usa_data = df[df["country"] == "United States"][["covid_deaths", "total_deaths"]].sum()

# Define labels and values
labels = ["COVID Deaths", "Other Deaths"]
values = [usa_data["covid_deaths"], usa_data["total_deaths"] - usa_data["covid_deaths"]]

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(values, labels=labels, autopct="%1.1f%%", colors=["red", "gray"], startangle=90, shadow=True)

# Title
plt.title("Proportion of COVID Deaths vs. Total Deaths in USA")

# Save plot
plt.savefig("D:/projects for cv/covid19-visualization/screenshots/covid_deaths_pie_chart.png")

# Show plot
plt.show()
```
